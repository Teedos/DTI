```python
import pickle
import json
```

```python
def load_pickle(path):
    with open(path,'rb') as f:
        my_dict = pickle.load(f)
    return my_dict
    
def save_dict(path,dict):
    with open(path,'wb') as handle:
        pickle.dump(dict,handle)
        
def load_dict(path):
    with open(path) as f:
        my_dict = json.load(f)
    return my_dict
```

```python
path='../../data//uniprotuni_seq2id.pkl'
uni_seq2id = load_pickle(path)
```

```python
path = '../../data/uniprot/uni_id2desc.pkl'
uni_id2desc = load_pickle(path)
```

```python
path = '../../data/yamanishi08/yama08_id2seq.pkl'
yama_id2seq = load_pickle(path)
seq_list = yama_id2seq.values()
path = '../../data/yamanishi08/yama08_seq2id.pkl'
yama_seq2id = load_pickle(path)
```

```python
yama_prot2desc = {}
for seq in seq_list:
    if seq in uni_seq2id:
        uni_id = uni_seq2id[seq]
        #if uni_id in uni_id2desc:
        yama_id = yama_seq2id[seq]
        if type(uni_id2desc[uni_id]) is float:
            yama_prot2desc[yama_id] = ''
        else:
            #print(uni_id2desc[uni_id])
            yama_prot2desc[yama_id] = uni_id2desc[uni_id]

for yama_id in yama_id2seq.keys():
    if yama_id not in yama_prot2desc:
        yama_prot2desc[yama_id] = ''
```

```python
count_null=0
for desc in yama_prot2desc.values():
    if desc == '':
        count_null +=1
```

```python
print(len(yama_prot2desc))
print(count_null)    ##57 proteins don't have description
```

```python
path = '../../yamanishi08/yama08_prot2desc.pkl'
save_dict(path,yama_prot2desc)
```
