```python
import pandas as pd
```

```python
import pickle
import json 
def load_pickle(path):
    with open(path,'rb') as handle:
        my_dict = pickle.load(handle)
    return my_dict

def save_pickle(path, my_dict):
    with open(path,'wb') as handle:
        pickle.dump(my_dict, handle)

def load_json_dict(path):
    with open(path) as f:
        my_dict = json.load(f)
    return my_dict
```

```python
path = '../../data/yamanishi08/split/data.csv'
data_df = pd.read_csv(path,sep=',')
data_df.head()
```

```python
path = '../../data/yamanishi08/yama_drug_id2smiles.pkl'
drug_id2smiles = load_pickle(path)
drug_smiles2id = {v:k for k,v in drug_id2smiles.items()}
```

```python
path = '../../data/yamanishi08/yama_prot_id2seq.pkl'
prot_id2seq = load_pickle(path)
prot_seq2id = {v:k for k,v in prot_id2seq.items()}
```

```python
data_df['compound_iso_smiles'] = data_df['compound_iso_smiles'].map(drug_smiles2id)
data_df['target_sequence'] = data_df['target_sequence'].map(prot_seq2id)
data_df.head()
```

```python
path = '../../data/yamanishi08/split/origin_data.csv'
data_df.to_csv(path,index=False)
```
