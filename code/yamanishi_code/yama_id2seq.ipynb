```python
import pandas as pd
df = pd.read_csv('../../data/yamanishi08/989proseq.csv')
seq_list = df['seq'].values.tolist()
id_list = df['pro_ids'].values.tolist()
```

```python
yama_id2seq = {}
for id, seq in zip(id_list, seq_list):
    yama_id2seq[id] = seq
```

```python
print(len(yama_id2seq))
inv_map = {v: k for k, v in yama_id2seq.items()}
print(len(inv_map))
```

```python
import pickle
def save_dict(path,dict):
    with open(path,'wb') as handle:
        pickle.dump(dict,handle)
```

```python
path = '../../yamanishi08/yama08_id2seq.pkl'
save_dict(path,yama_id2seq)
```

```python
path = '../../yamanishi08/yama08_seq2id.pkl'
yama08_seq2id = {v: k for k, v in yama_id2seq.items()}
save_dict(path,yama08_seq2id)
print(yama08_seq2id)
```
