```python
import pickle
```

```python
def load_pickle(path):
    with open(path,'rb') as handle:
        my_dict = pickle.load(handle)
    return my_dict

def save_pickle(path, my_dict):
    with open(path,'wb') as handle:
        pickle.dump(my_dict, handle)
```

```python
####load binding smiles to canonic smiles
path = '../../data/BindingDB/BD_smi2cansmi.pkl'
binding_smiles2can = load_pickle(path)
```

```python
print(len(binding_smiles2can))
```

```python
#####load drug bank ids to canonical smiles
path = '../../data/BindingDB/DBid2cansmiles.pkl'
db_id2can = load_pickle(path)
```

```python
###### reverse map for canonical smiles to drug bank id
db_can2id = {v: k for k, v in db_id2can.items()}
path = '../../data/BindingDB/db_can_smiles2id.pkl'
save_pickle(path, db_can2id)
```

```python
####load db id to description
path = '../../data/drugbank/db2desc.pkl'
db_id2desc = load_pickle(path)
```

```python
unique_smiles = binding_smiles2can.values()
print(len(unique_smiles))
```

```python
#######create binding db canonical smiles to description
binding_can_smiles2desc = {}
count_in_db = 0
count_null = 0
for smiles in unique_smiles:
    if smiles in db_can2id:
        count_in_db +=1
        db_id = db_can2id[smiles]
        desc = db_id2desc[db_id]
        if desc =='':
            count_null +=1
        binding_can_smiles2desc[smiles] = desc
    else:
        binding_can_smiles2desc[smiles] = ''
        count_null +=1
```

```python
#####save binding canonical to description
path = '../../data/BindingDB/binding_can_smiles2desc.pkl'
save_pickle(path,binding_can_smiles2desc)
```

```python
for v,k in binding_can_smiles2desc.items():
    print(v,k)
```

```python
print(len(binding_can_smiles2desc))
print('successfully linked', count_in_db)
print('Do not have description',count_null)
```
